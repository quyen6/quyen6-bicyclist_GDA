```python
import numpy as np
import pandas as pd
```

```python
bi_2019_q1 = pd.read_csv('C:/Users/MyQuyen/Downloads/data_bicycle/Divvy_Trips_2019_Q1.csv')
bi_2020_q1 = pd.read_csv('C:/Users/MyQuyen/Downloads/data_bicycle/Divvy_Trips_2020_Q1.csv')
```

```python
# đổi tên cho đồng nhất
bi_2019_q1 = bi_2019_q1.rename(columns={
    'trip_id': 'ride_id',
    'start_time': 'started_at',
    'end_time': 'ended_at',
    'from_station_name': 'start_station_name',
    'from_station_id': 'start_station_id',
    'to_station_name': 'end_station_name',
    'to_station_id': 'end_station_id',
    'usertype': 'member_casual'
})
```

```python
# chuyển đổi format ride_id
bi_2019_q1['ride_id'] = bi_2019_q1['ride_id'].astype(str)
```

```python
#  Xác định các cột chung xuất hiện ở TẤT CẢ các bảng
common_columns = [
    'ride_id', 'started_at', 'ended_at', 
    'start_station_name', 'start_station_id', 
    'end_station_name', 'end_station_id', 
    'member_casual', 'ride_length', 'day_of_week'
]
```

```python
# Lọc các DataFrame chỉ lấy các cột chung này
bi_2019_q1_clean = bi_2019_q1[common_columns]
bi_2020_q1_clean = bi_2020_q1[common_columns]
```

```python
# Gộp các bảng theo hàng dọc (Concatenate)
data = pd.concat([bi_2019_q1_clean, bi_2020_q1_clean], ignore_index=True)
```

```python
data.head()
```

```python
# chuẩn hóa cột member_casual 
data['member_casual'] = data['member_casual'].replace({'Subscriber': 'member', 'Customer': 'casual'})
```

```python
data.head()
```

```python
data.info()
```

```python
## tìm giá trị mà ride_length < 0
# 1. Ép kiểu cột sang ride_length dạng Timedelta (Khoảng thời gian) 
data['ride_length'] = pd.to_timedelta(data['ride_length'], errors='coerce')

# 2. Chuyển đổi thành SỐ PHÚT (hoặc số giây bằng cách thay 'm' bằng 's') để dễ tính toán
data['ride_length'] = (data['ride_length'].dt.total_seconds() / 60).round(2)
```

```python
data.info()
```

```python
## đếm giá trị mà ride_length < 0
(data['ride_length']<0).value_counts()
```

```python
# Lọc lấy các chuyến đi hợp lệ từ 1 phút
data_clean = data[(data['ride_length'] >= 1) & (data['ride_length'] <= 1440)]

print(data_clean['ride_length'].describe())
```

```python
data_clean.head()
```

```python
# Tính trung bình (mean) của thời gian đi xe (ride_length)
data_clean = data[(data['ride_length'] >= 1) & (data['ride_length'] <= 1440)].copy()

data_clean['started_at'] = pd.to_datetime(data_clean['started_at'])
data_clean['year'] = data_clean['started_at'].dt.year

# Tính lại mean
print(data_clean.groupby('year')['ride_length'].mean())
```

```python
# Tính tổng số chuyến, trung bình, trung vị, max, min của từng nhóm
print(data_clean.groupby('member_casual')['ride_length'].agg(['count', 'mean', 'median', 'max']))
```

```python
# Tính thời gian đi xe trung bình của 2 nhóm theo từng ngày trong tuần
print(data_clean.groupby(['member_casual', 'day_of_week'])['ride_length'].mean())

# Đếm tổng số chuyến đi của 2 nhóm theo từng ngày trong tuần
print(data_clean.groupby(['member_casual', 'day_of_week'])['ride_id'].count())
```

```python
data_clean.to_csv('cyclistic_summary_by_day.csv', index=False)
print("Đã xuất file summary thành công!")
```
